```r
library('MASS')
```

```r
#reading the dataset in R 

dataset<-read.csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

```r
dataset
```

```r
#ploting the dataset in corrleogram

plot(dataset)
```

```r
install.packages('corrplot')
```

```r
#infering the correlation 
cr<-cor(dataset)

corrplot::corrplot(cr,type='lower')
```

```r
corrplot::corrplot(cr,method='number')
```

```r
#divide the dataset into train and test
split=sample(2,nrow(dataset),prob = c(0.8,0.2),replace = T)
```

```r
train<-dataset[split==1,]
test<-dataset[split==2,]
```

```r
#creating our linear regression model with medv as independent vairable and others as dependent variable

model<-lm(medv~.,data=train)
#summary of the model 

summary(model)
```

```r
#predicitng the test values 
pred<-predict(model,test)
```

```r
#ploting the actual vs the predeicted data
plot(pred,type='l',lty=1.8,col='blue')
lines(test$medv,type='l',col='green')
```

```r
#ploting the actual vs the predeicted data
plot(pred,col='red')
points(test$medv,col='green')
```

```r
#we can clearly see that indus and age are not correlated with medv 
#THUS removing the same and recreating the model
model<-lm(medv~.-indus-age,data = train)

summary(model)

```

```r
#predicitng the test values 
pred<-predict(model,test)
```

```r
#ploting the actual vs the predeicted data
plot(pred,type='l',lty=1.8,col='blue')
lines(test$medv,type='l',col='green')
```

```r
#ploting the actual vs the predeicted data
plot(pred,col='red')
points(test$medv,col='green')
```
